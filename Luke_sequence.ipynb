# В математике, последовательностями Люка называют семейство пар линейных рекуррентных последовательностей второго порядка, впервые рассмотренных Эдуардом Люка.

# Напишите функцию luka(L0, L1, n), которая принимает на вход параметры:

L0, L1 - 0й и 1й члены последовательности соответственно
n - номер числа из последовательности, которое необходимо вернуть

```python
def luka(L0, L1, n):
    if n == 0:
        return L0
    elif n == 1:
        return L1
    else:
        for i in range(n - 1):
            L0, L1 = L1, L0 + L1
        return L1
```

```python
def luka(L0, L1, n):
    for _ in range(n):
        L0, L1 = L1, L0 + L1
    return L0
```

```python
L0, L1, n = 2, 1, 180
luka(L0, L1, n)
```

# Напишите функцию fi(L0, L1, n) такую что:

* L0, L1 - 0й и 1й члены последовательности Люка соответственно
* n - номер числа из последовательности, которое необходимо вернуть
* Возвращаемое значение - это отношение 2 членов последовательности: Ln / Ln-1
 
Гарантируется, что n≥2

В данной задаче нам очень важна точность вычислений. Поэтому мы будем использовать необычный тип данных - Decimal.

```python
from decimal import Decimal
getcontext().prec = 50

def fi(L0, L1, n):
    L1 = Decimal(L1)
    for _ in range(n-1):
        L0, L1 = L1, L0 + L1
    return L1 / L0
```

```python
L0, L1, n = 0, 1, 11
fi(L0, L1, n)
```

# Напишите функцию super_L(n), находящую n-е число Люка.

Используйте формулы для разложения из предыдущего шага, чтобы увеличить быстродействие вашей функции.

```python
import time
```

```python
def super_L(n):
    if n % 5 == 0:
        n //= 5
        return super_L(n) ** 5 - 5 * ((-1) ** n) * super_L(n) ** 3 + 5 * super_L(n)
    elif n % 3 == 0:
        n //= 3
        return super_L(n) ** 3 - 3 * ((-1) ** n) * super_L(n)
    elif n % 2 == 0:
        n //= 2
        return super_L(n) ** 2 - 2 * ((-1) ** n)
    else:
        L0 = 2
        L1 = 1
        for _ in range(n):
            L0, L1 = L1, L0 + L1
        return L0
```

```python
def L(n):
    L0 = 2
    L1 = 1
    for _ in range(n):
        L0, L1 = L1, L0 + L1
    return L0
```

```python
start_time = time.time()
super_L(122148)
print("--- %s seconds ---" % (time.time() - start_time))
start_time = time.time()
L(122148)
print("--- %s seconds ---" % (time.time() - start_time))
```
